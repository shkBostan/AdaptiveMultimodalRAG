# Text Embedding Demo

This notebook demonstrates text embedding generation using BERT.


```python
import sys
sys.path.append('../')

from embeddings.bert_embedding import BERTEmbedding
import numpy as np
```

```python
# Initialize BERT embedding model
bert_model = BERTEmbedding(model_name='bert-base-uncased')
bert_model.load_model()
```
